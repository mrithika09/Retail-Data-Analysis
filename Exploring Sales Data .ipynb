## What time should we display advertisements to maximise the likelihood of customer's buying product?

```python
df.head()
```

```python
df['Time'] = df['Order Date'].dt.hour
df.head()
```

```python
bdf = df.groupby("Time").agg(Total_Count=("Time", "count")).reset_index()
bdf
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Time', y='Total_Count', data=bdf)
plt.title("Best Time for Advertisement", color="red")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(bdf['Time'], bdf['Total_Count'], marker='o')
plt.title("Best Time for Advertisement", color="red")
plt.xticks(np.arange(1, 24))
plt.grid()
plt.show()
```
